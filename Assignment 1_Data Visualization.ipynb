### Lakukan Data Visualization menggunakan datasets titanic.csv untuk menganalisa informasi Single Variable, Two Variable dan Multi Variable!

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
data = pd.read_csv("datasets/titanic.csv")
data
```

```python
data.describe()
```

### Single Variable

```python
data['Survived'].hist()
plt.show()
```

```python
data['Pclass'].hist()
plt.show()
```

```python
data['Sex'].hist()
plt.show()
```

```python
data['Age'].hist()
plt.show()
```

```python
sns.distplot(data['Age'])
plt.show()
```

```python
data['Siblings/Spouses Aboard'].hist()
plt.show()
```

```python
data['Parents/Children Aboard'].hist()
plt.show()
```

```python
data['Fare'].hist()
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
sns.pairplot(data, hue='Survived')
plt.show()
```

```python
sns.heatmap(data.corr(), annot=True)
plt.show()
```

### Two Variables

```python
data.hist("Survived", by="Pclass", grid="False", layout=[1,3],figsize = [10,3])
```

```python
data.hist("Survived", by="Sex",figsize = [10,3])
```

### Multi Variables

```python
data.hist("Age", bins=10, by = ["Survived", "Sex"], layout=[1,4],figsize = [20,3])
```
